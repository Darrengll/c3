```python
P0list=Plist[:,0]
P1list=Plist[:,1]
P2list=Plist[:,2]
```

```python
def T1_fit(t, a, T1, b):
    return a*np.exp(-t/T1)+b

init_guess = [1, 20e-6, 0]

solution, cov = curve_fit(
    T1_fit, tlist, P1list, p0=init_guess
)
a, T1, b = solution
```

```python
fig,ax=plt.subplots()
ax.scatter(tlist,P1list)
ax.plot(tlist,T1_fit(tlist, a, T1, b),'r')
ax.set_title("T1={:.3f}us".format(T1*1e6))
```

```python
seqs = []
for id_num in id_num_list:
    seq = qt_utils.ramsey_sequence(length=id_num, target=0)
    seqs.append(seq)
pops = single_qubit_experiment(seqs,exp=exp)
Plist=np.array([pop.numpy().flatten().tolist() for pop in pops])
```

```python
P0list=Plist[:,0]
P1list=Plist[:,1]
P2list=Plist[:,2]
```

```python
def T2_fit(x, T2, freq, phase, amp, e):
    return amp * np.exp(-x / T2) * np.cos(2 * np.pi * x * freq + phase) + e

init_guess = [0.4e-6, 50e6, 0,0.5,0.5]

solution, cov = curve_fit(
    T2_fit, tlist, P1list, p0=init_guess
)
T2, freq, phase, amp, e = solution
```
